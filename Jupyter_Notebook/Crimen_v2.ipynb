```python
from math import dist

Estacion = []
Lat = []
Lon = []

for j in trange(len(crimen)):
    
    a = (crimen["Longitud"][j], crimen["Latitud"][j])
    distancias = []
    for i in range(len(cat_estacion)):
        b = (cat_estacion["longitud"][i], cat_estacion["latitud"][i])
        distancias.append(dist(a,b))
        #print(str(cat_estacion.index[i]) + " " + str(cat_estacion["longitud"][i]))
    #b = (4, 5, 6)

    min_index=distancias.index(min(distancias))
    Estacion.append(cat_estacion["nom_estac"].iloc[min_index])
    Lat.append(cat_estacion["latitud"].iloc[min_index])
    Lon.append(cat_estacion["longitud"].iloc[min_index])
    


#print(distancias)
#print(cat_estacion["nom_estac"].iloc[min_index])
#print(cat_estacion["longitud"].iloc[min_index])
#print(cat_estacion["latitud"].iloc[min_index])
#print(dist(a,b))
```

```python
crimen2["Estacion"] = Estacion
crimen2["Longitud"] = Lon
crimen2["Latitud"] = Lat

crimen2
```

```python
crimen2 = crimen2.drop(['Calle'], axis=1)
```
