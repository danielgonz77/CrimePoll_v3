````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
import plotly.express as px
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import math
````

````python
df_std = pd.read_csv('../CSV_limpios/crime_air_std.csv')
````

````python
df_std = df_std.drop(df_std.columns[[0, -1, -2]], axis=1)
````

````python
df_std = df_std[['ID', 'Dia', 'Mes', 'Anio', 'Latitud', 'Longitud', 'Alcaldia', 'Alcaldia_num', 'AQI', 'Crimenes', 'Crime_cat', 'Crime_cat_num']]
````

````python
df_std
````

````python
df_std["Crimenes"].nunique()
````

````python
#categories = [math.floor(df_group_by["% / 10 AQI"].min()), 0, 0.6, 1, math.ceil(df_group_by["% / 10 AQI"].max())]
categories = [i+1 for i in range(0, math.ceil(df_std["Crimenes"].max()+4), 5)]

labels = ["{0} - {1}".format(i, i+4) for i in range(1, math.ceil(df_std["Crimenes"].max()), 5)]
labels2 = list(range(1, len(labels)+1))
print(labels)
print(categories)
labels2
````

````python
df_std["Crime_cat"] = pd.cut(df_std['Crimenes'], bins=categories, right=False, labels=labels)
df_std["Crime_cat_num"] = pd.cut(df_std['Crimenes'], bins=categories, right=False, labels=labels2)
````

````python
df_train, df_test = train_test_split(df_std, test_size=0.2)
````

````python
X_train = df_train.drop(df_train.columns[[0, 1, 2, 3, 4, 5, 6, -1, -2, -3]], axis=1).values
y_train = df_train["Crime_cat_num"].values #target
````

````python
X_train
````

# Using Cross-Validation to evaluate our model

````python
from sklearn.model_selection import cross_val_score, cross_val_predict
````

````python
model = LogisticRegression(max_iter=1000)

#La regresión logística, a pesar de su nombre, es un modelo lineal de clasificación más que de regresión
````

````python
accuracies = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy")
````

````python
np.mean(accuracies)
````

# Model Tuning

What is model tuning?

Model tuning is try to determinate the parameter of the model (these are also know as "hyperparameters") that maximaze the model performance.

````python
for reg_param in (0.1, 0.3, 0.9, 1, 1.2, 1.5, 1.7, 2):
    print(reg_param)
    model = LogisticRegression(max_iter=5000, C=reg_param)
    accuracies = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy")
    print(f"Accuracy: {np.mean(accuracies) * 100:.2f}%")
````

````python
for reg_param in (1.1, 1.2, 1.3, 1.4):
    print(reg_param)
    model = LogisticRegression(max_iter=7000, C=reg_param)
    accuracies = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy")
    print(f"Accuracy: {np.mean(accuracies) * 100:.2f}%")
````

### Final model

````python
model = LogisticRegression(max_iter=7000, C=1.3)
accuracies = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy")
print(f"Accuracy: {np.mean(accuracies) * 100:.2f}%")
````

## How well does our model do on the Test Set?

````python
X_test = df_test.drop(df_test.columns[[0, 1, 2, 3, 4, 5, 6, -1, -2, -3]], axis=1).values
y_test = df_test["Crime_cat_num"].values #target
````

### Train our final model using our full Training Dataset

````python
model.fit(X_train, y_train)
````

````python
y_test_pred = model.predict(X_test)
````

````python
test_set_correctly_classified = y_test_pred == y_test
test_set_accuracy = np.mean(test_set_correctly_classified)
````

````python
print(f"Test set accuracy: {test_set_accuracy * 100:.2f}")
````

````python
df_predictions_test = df_test.copy()
df_predictions_test["correct_predict"] = test_set_correctly_classified
df_predictions_test["predictions"] = y_test_pred 
df_predictions_test["prediction_label"] = df_predictions_test["predictions"].map({1:"1 - 5", 2:"6 - 10", 3:"11 - 15", 4:"16 - 20", 5:"21 - 25", 6:"26 - 30", 7:"31 - 35", 8:"36 - 40", 9:"41 - 45", 10:"46 - 50", 11:"51 - 55", 12:"56 - 60"})
````
