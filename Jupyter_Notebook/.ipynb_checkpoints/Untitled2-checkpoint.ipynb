```python
estado = {}
latitud = {}
longitud = {}
for i in range(len(df_estacion)):
    latitud[df_estacion["nom_estac"][i]] = df_estacion["latitud"][i]
    longitud[df_estacion["nom_estac"][i]] = df_estacion["longitud"][i]
    address = geo.reverse(str(df_estacion["latitud"][i]) + ", " + str(df_estacion["longitud"][i]))
    #print(address, address.latitude, address.longitude)
    #x = str(address).split(", ")
    Estado_1 = ""
    if "Ciudad de México" in str(address):
        Estado_1 = "Ciudad de México"
        #print(Estado)
    else:
        Estado_1 = "Estado de México"
    estado[df_estacion["nom_estac"][i]] = Estado_1
```

```python
estado_toAdd = []
latitud_toAdd = []
longitud_toAdd = []
#estado[df_aire_100['Lugar'][0]]
for i in range(len(df_aire)):
    estado_toAdd.append(estado[df_aire['Lugar'][i]])
    latitud_toAdd.append(latitud[df_aire['Lugar'][i]])
    longitud_toAdd.append(longitud[df_aire['Lugar'][i]])
```
