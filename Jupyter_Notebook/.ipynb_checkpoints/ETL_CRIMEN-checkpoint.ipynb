```python
#se importan los datos desde google drive
from google.colab import drive

drive.mount('/content/drive')

path = "/content/drive/MyDrive/CrimePoll/carpetas_ss_junio2020.csv"
df_1 = pd.read_csv(path)

path = "/content/drive/MyDrive/CrimePoll/carpetas_completa_marzo_2022.csv"
df_2 = pd.read_csv(path)
```

```python
#se comprueba el contenio de los data frames
# df_2
df_1
```

```python
#separamos la fecha_hechos en dia, mes, anio, hora del primer dataframe
df_1['fecha_hechos'] = pd.to_datetime(df_1['fecha_hechos'])
df_1['fecha_hechos'].dtype
# Reset our index so datetime_utc becomes a column
# df_1.reset_index(inplace=True)

# Create new columns
df_1['dia'] = df_1['fecha_hechos'].dt.day
df_1['mes'] = df_1['fecha_hechos'].dt.month
df_1['anio'] = df_1['fecha_hechos'].dt.year
df_1['hora'] = df_1['fecha_hechos'].dt.hour

# df_1
```

```python
#separamos la fecha_hechos en dia, mes, anio, hora del primer dataframe
df_2['fecha_hechos'] = pd.to_datetime(df_2['fecha_hechos'],errors = 'coerce')

# Create new columns
df_2['dia'] = df_2['fecha_hechos'].dt.day
df_2['mes'] = df_2['fecha_hechos'].dt.month
df_2['anio'] = df_2['fecha_hechos'].dt.year
df_2['hora'] = df_2['fecha_hechos'].dt.hour
```

```python
#creamos un dataframe a partir de las series en las que juntamos la informacion de los dos dataframes
frames = [df_1, df_2]
df_3 = pd.concat(frames)
df = pd.DataFrame()
df['Dia'] = df_3['dia']
df['Mes'] = df_3['mes'] 
df['Anio'] = df_3['anio'] 
df['Hora'] = df_3['hora'] 
df['Delito'] = df_3['delito'] 
df['Categoria'] = df_3['categoria_delito'] 
df['Alcaldia'] = df_3['alcaldia_hechos']
df['Longitud'] = df_3['longitud']
df['Latitud'] = df_3['latitud']
```

```python
df = df.dropna(subset=['Dia', 'Delito','Alcaldia'])
df['Dia'] = df['Dia'].astype(int)
df['Mes'] = df['Mes'].astype(int)
df['Anio'] = df['Anio'].astype(int)
df['Hora'] = df['Hora'].astype(int)
```

```python
alcaldias = ['AZCAPOTZALCO', 'ALVARO OBREGON', 'BENITO JUAREZ', 'COYOACAN', 'CUAJIMALPA DE MORELOS', 'CUAUHTEMOC',
             'GUSTAVO A MADERO', 'IZTACALCO', 'IZTAPALAPA', 'LA MAGDALENA CONTRERAS', 'MIGUEL HIDALGO', 'MILPA ALTA', 'TLALPAN', 'TLAHUAC', 'VENUSTIANO CARRANZA', 'XOCHIMILCO']

df = df[df['Alcaldia'].isin(alcaldias)]

delitos = ['ABUSO SEXUAL', 'HOMICIDIO POR ARMA DE FUEGO', 'ROBO A CASA HABITACION SIN VIOLENCIA', 'ROBO A NEGOCIO CON VIOLENCIA',
           'ROBO A NEGOCIO SIN VIOLENCIA', 'ROBO A TRANSEUNTE EN VIA PUBLICA CON VIOLENCIA',
           'VIOLACION', 'VIOLENCIA FAMILIAR', 'ROBO A TRANSEUNTE DE CELULAR CON VIOLENCIA', 'ROBO A TRANSEUNTE DE CELULAR SIN VIOLENCIA', 
           'ROBO A PASAJERO A BORDO DE METRO SIN VIOLENCIA',
           'ROBO A TRANSEUNTE EN VIA PUBLICA SIN VIOLENCIA', 'ROBO A REPARTIDOR CON VIOLENCIA',
           'ROBO A PASAJERO A BORDO DE TRANSPORTE PÚBLICO CON VIOLENCIA']
           
df = df[df['Delito'].isin(delitos)]
df
```

```python
#se borran nulos
df = df.dropna(subset=['Longitud'])
```

```python
#se comprueba los nulos
df.isnull().sum()
```

```python
# categorias

#con violencia: 
violencia = ['ABUSO SEXUAL', 'HOMICIDIO POR ARMA DE FUEGO', 'ROBO A NEGOCIO CON VIOLENCIA', 'ROBO A TRANSEUNTE EN VIA PUBLICA CON VIOLENCIA', 'VIOLACION',
             'VIOLENCIA FAMILIAR', 'ROBO A TRANSEUNTE DE CELULAR CON VIOLENCIA', 'ROBO A REPARTIDOR CON VIOLENCIA', 'ROBO A PASAJERO A BORDO DE TRANSPORTE PÚBLICO CON VIOLENCIA']
#sin violencia
sinViolencia = ['ROBO A CASA HABITACION SIN VIOLENCIA', 'ROBO A NEGOCIO SIN VIOLENCIA', 'ROBO A TRANSEUNTE DE CELULAR SIN VIOLENCIA',
                'ROBO A PASAJERO A BORDO DE METRO SIN VIOLENCIA', 'ROBO A TRANSEUNTE EN VIA PUBLICA SIN VIOLENCIA']

violencia

df['Categoria'] = df['Delito'].apply(lambda x: "CON VIOLENCIA" if x in violencia else "SIN VIOLENCIA")
```

```python
anios = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
df = df[df['Anio'].isin(anios)]
```
