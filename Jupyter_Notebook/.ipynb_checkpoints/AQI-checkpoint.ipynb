```python
import aqi
list_AQI = []

for i in range(len(df_aire_cdmx)):
    if(df_aire_cdmx["Particula"].values[i] == "PM2.5"):
        if((df_aire_cdmx["Valor"].values[i] != -99) & (df_aire_cdmx["Valor"].values[i] < 500.4)):
            myaqi = aqi.to_iaqi(aqi.POLLUTANT_PM25, df_aire_cdmx["Valor"].values[i], algo=aqi.ALGO_EPA)
            list_AQI.append(myaqi)
        else:
            list_AQI.append(-99)
    elif(df_aire_cdmx["Particula"].values[i] == "PM10"):
        if((df_aire_cdmx["Valor"].values[i] != -99) & (df_aire_cdmx["Valor"].values[i] < 604)):
            #print(df_aire_cdmx["Valor"].values[i])
            myaqi = aqi.to_iaqi(aqi.POLLUTANT_PM10, df_aire_cdmx["Valor"].values[i], algo=aqi.ALGO_EPA)
            list_AQI.append(myaqi)
        else:
            list_AQI.append(-99)
    elif(df_aire_cdmx["Particula"].values[i] == "CO"):
        if((df_aire_cdmx["Valor"].values[i] != -99) & (df_aire_cdmx["Valor"].values[i] < 50.4)):
            myaqi = aqi.to_iaqi(aqi.POLLUTANT_CO_8H, df_aire_cdmx["Valor"].values[i], algo=aqi.ALGO_EPA)
            list_AQI.append(myaqi)
        else:
            list_AQI.append(-99)
    else:
        list_AQI.append(-99)
```

```python
df_aire_cdmx['AQI'] = list_AQI
```
