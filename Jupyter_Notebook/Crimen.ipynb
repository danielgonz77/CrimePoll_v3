```python
crimen_v2 = pd.read_csv("E:\\7mo\\TT1\\Datos_crimen\\carpetas_completa_marzo_2022.csv")
df_2 = pd.DataFrame(crimen_v2)
```

```python
df_2
```

```python
#separamos la fecha_hechos en dia, mes, anio, hora del primer dataframe
df_2['fecha_hechos'] = pd.to_datetime(df_2['fecha_hechos'],errors = 'coerce')

# Create new columns
df_2['dia'] = df_2['fecha_hechos'].dt.day
df_2['mes'] = df_2['fecha_hechos'].dt.month
df_2['anio'] = df_2['fecha_hechos'].dt.year
df_2['hora'] = df_2['fecha_hechos'].dt.hour
```

```python
#creamos un dataframe a partir de las series en las que juntamos la informacion de los dos dataframes

df_3 = df_2
df = pd.DataFrame()
df['Dia'] = df_3['dia']
df['Mes'] = df_3['mes'] 
df['Anio'] = df_3['anio'] 
df['Hora'] = df_3['hora'] 
df['Delito'] = df_3['delito'] 
df['Categoria'] = df_3['categoria_delito'] 
df['Calle'] = df_3['calle_hechos']
df['Colonia'] = df_3['colonia_hechos']
df['Alcaldia'] = df_3['alcaldia_hechos']
df['Longitud'] = df_3['longitud']
df['Latitud'] = df_3['latitud']
```

```python
df
```

```python
df = df.dropna(subset=['Dia', 'Delito','Alcaldia'])
df['Dia'] = df['Dia'].astype(int)
df['Mes'] = df['Mes'].astype(int)
df['Anio'] = df['Anio'].astype(int)
df['Hora'] = df['Hora'].astype(int)
```

```python
df
```

```python
alcaldias = ['AZCAPOTZALCO', 'ALVARO OBREGON', 'BENITO JUAREZ', 'COYOACAN', 'CUAJIMALPA DE MORELOS', 'CUAUHTEMOC',
             'GUSTAVO A MADERO', 'IZTACALCO', 'IZTAPALAPA', 'LA MAGDALENA CONTRERAS', 'MIGUEL HIDALGO', 'MILPA ALTA', 'TLALPAN', 'TLAHUAC', 'VENUSTIANO CARRANZA', 'XOCHIMILCO']

df = df[df['Alcaldia'].isin(alcaldias)]

delitos = ['ABUSO SEXUAL', 'HOMICIDIO POR ARMA DE FUEGO', 'ROBO A CASA HABITACION SIN VIOLENCIA', 'ROBO A NEGOCIO CON VIOLENCIA',
           'ROBO A NEGOCIO SIN VIOLENCIA', 'ROBO A TRANSEUNTE EN VIA PUBLICA CON VIOLENCIA',
           'VIOLACION', 'VIOLENCIA FAMILIAR', 'ROBO A TRANSEUNTE DE CELULAR CON VIOLENCIA', 'ROBO A TRANSEUNTE DE CELULAR SIN VIOLENCIA', 
           'ROBO A PASAJERO A BORDO DE METRO SIN VIOLENCIA',
           'ROBO A TRANSEUNTE EN VIA PUBLICA SIN VIOLENCIA', 'ROBO A REPARTIDOR CON VIOLENCIA',
           'ROBO A PASAJERO A BORDO DE TRANSPORTE PÚBLICO CON VIOLENCIA']
           
df = df[df['Delito'].isin(delitos)]
df
```
